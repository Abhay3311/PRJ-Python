```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
resultData = pd.read_csv("https://raw.githubusercontent.com/Anpr1211/DOE_Datasets/main/data_viz.csv")
resultData.head()
```

```python
resultDatapv = pd.pivot_table(resultData,index=["Stream"],values=['StudentId'],aggfunc={'StudentId':"count"})
resultDatapv.reset_index(inplace=True)
resultDatapv
```

```python
plt.bar(x=resultDatapv["Stream"],height=resultDatapv["StudentId"])

plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Optional: To make the figure larger
sns.barplot(x='Stream', y='StudentId', data=resultDatapv, hue='Stream', dodge=False, palette='husl')

# Add the title and labels
plt.title('Number of Students per Stream')
plt.xlabel('Stream')
plt.ylabel('Number of Students')

# Display the chart
plt.show()
```

```python
colors = ['red', 'green', 'blue']
plt.bar(x=resultDatapv['Stream'],height=resultDatapv['StudentId'], color=colors)
plt.xlabel("Stream")
plt.ylabel("Count of Student Id")
plt.title("Numbers of students from different streams that have attempted the exam")
plt.xticks(rotation= 45)
plt.show()
```

Create a pie chart of the gender of students who have attempted the exam.

```python

resultGender =pd.pivot_table(resultData,index=['Gender'],values=['StudentId'],aggfunc={'StudentId': 'count'})
resultGender.reset_index(inplace=True)
resultGender
```

```python
color = ["#F75B95", "#06B0F0"]
plt.pie(x=resultGender['StudentId'],labels=resultGender['Gender'].str.title(),colors = color)
plt.title('Share of Gender')
plt.show()
```

```python
color = ["#00ffff", "#ffff00"]
plt.pie(x=resultGender['StudentId'],labels=resultGender['Gender'].str.title(),colors = color,autopct= '%.0f%%')
plt.title('Pie Chart of the gender of students')
plt.show()
```

```python
SalesData =pd.read_csv("https://raw.githubusercontent.com/Anpr1211/DOE_Datasets/main/%5BPractiseExercise-1%5D%20B8.2%20All-in-One%20Furnishings%20Percentage%20Analysis%20-%20Sales_data.csv")
SalesData.head()
```

```python
beastData = pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE/main/MrBeast_youtube_stat.csv")
beastData.head()
```

```python
beastPivot =pd.pivot_table(beastData,index=['year'],values=['viewCount'],aggfunc={'viewCount':"sum"}).reset_index()
beastPivot
```

```python
beastPivot1 =pd.pivot_table(beastData,index=['year'],values=['impressionCount'],aggfunc={'impressionCount':"sum"}).reset_index()
beastPivot1
```

```python
plt.plot(beastPivot['year'],beastPivot['viewCount'],label="viewCount")
plt.plot(beastPivot1['year'],beastPivot1['impressionCount'],label="impressionCount")

plt.xlabel("viewCount")
plt.ylabel("viewCount and impressionCount")
plt.title("Year wise Likes and Subscribers ")
plt.legend()
plt.show()
```

```python
plt.scatter(beastData.query('kind_stats == "Shorts"')['impressionCount'],beastData.query('kind_stats == "Shorts"')['viewCount'])
plt.show()
```

```python
plt.scatter(beastData.query('kind_stats == "Shorts"')['impressionCount'],beastData.query('kind_stats == "Shorts"')['viewCount'])
plt.scatter(beastData.query('kind_stats == "Long"')['viewCount'],beastData.query('kind_stats == "Long"')['viewCount'])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=beastData['impressionCount'], y=beastData['viewCount'])
plt.show()
```

Scattered plot with a regression line!

```python
plt.figure(figsize=(10,5))
sns.regplot(x=beastData['impressionCount'], y=beastData['viewCount'])
plt.show()
```

```python

plt.figure(figsize=(10,5))
sns.scatterplot(x=beastData['impressionCount'], y=beastData['viewCount'],hue=beastData['kind_stats'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lmplot(x='impressionCount', y='viewCount',hue='kind_stats',data=beastData)
plt.show()
```

```python
plt.figure(figsize=(10,6));
sns.swarmplot(x=beastData['kind_stats'], y=beastData['impressionCount'])
plt.show()
```
